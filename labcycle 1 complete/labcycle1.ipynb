```python
# Q3 View the individual channels of an RGB image.
import cv2
import numpy as np

# Load the color image
color_image = cv2.imread(r'C:\Users\2002a\Downloads\color.jpg')

# Check if image is loaded successfully
if color_image is None:
    print("Error: Unable to load image.")
else:
    # Split the image into its RGB channels
    blue_channel, green_channel, red_channel = cv2.split(color_image)

    # Create blank channel
    blank_channel = np.zeros_like(blue_channel)

    # Merge channels to create images representing each color channel
    blue_image = cv2.merge([blue_channel, blank_channel, blank_channel])
    green_image = cv2.merge([blank_channel, green_channel, blank_channel])
    red_image = cv2.merge([blank_channel, blank_channel, red_channel])

    # Create windows to display images
    cv2.namedWindow('Color Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Blue Channel', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Green Channel', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Red Channel', cv2.WINDOW_NORMAL)

    # Resize windows for better display
    display_width = 400
    display_height = 400
    cv2.resizeWindow('Color Image', display_width, display_height)
    cv2.resizeWindow('Blue Channel', display_width, display_height)
    cv2.resizeWindow('Green Channel', display_width, display_height)
    cv2.resizeWindow('Red Channel', display_width, display_height)

    # Display the images
    cv2.imshow('Color Image', color_image)
    cv2.imshow('Blue Channel', blue_image)
    cv2.imshow('Green Channel', green_image)
    cv2.imshow('Red Channel', red_image)


    # Wait for any key to be pressed to close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Q4 Perform image translations, rotations, flipping
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\2002a\Downloads\color.jpg')

# Translation
rows, cols = image.shape[:2]
translation_matrix = np.float32([[1, 0, 50], [0, 1, 30]])  # Translate 50 pixels to the right and 30 pixels down
translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))

# Rotation
rotation_matrix_left = cv2.getRotationMatrix2D((cols / 2, rows / 2), 90, 1)
rotated_image_left = cv2.warpAffine(image, rotation_matrix_left, (cols, rows))

rotation_matrix_right = cv2.getRotationMatrix2D((cols / 2, rows / 2), -90, 1)
rotated_image_right = cv2.warpAffine(image, rotation_matrix_right, (cols, rows))

rotation_matrix_top = cv2.getRotationMatrix2D((cols / 2, rows / 2), 180, 1)
rotated_image_top = cv2.warpAffine(image, rotation_matrix_top, (cols, rows))

rotated_image_bottom = image

# Flipping
horizontal_flip = cv2.flip(image, 1)
vertical_flip = cv2.flip(image, 0)
both_flip = cv2.flip(image, -1)

# Display the images
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.show()

plt.imshow(cv2.cvtColor(rotated_image_left, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image Right')
plt.show()

plt.imshow(cv2.cvtColor(rotated_image_right, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image Lefft')
plt.show()

plt.imshow(cv2.cvtColor(rotated_image_top, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image Top')
plt.show()

plt.imshow(cv2.cvtColor(rotated_image_bottom, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image Bottom')
plt.show()

plt.imshow(cv2.cvtColor(horizontal_flip, cv2.COLOR_BGR2RGB))
plt.title('Horizontal Flip')
plt.show()

plt.imshow(cv2.cvtColor(vertical_flip, cv2.COLOR_BGR2RGB))
plt.title('Vertical Flip')
plt.show()

plt.imshow(cv2.cvtColor(both_flip, cv2.COLOR_BGR2RGB))
plt.title('Both Flip')
plt.show()
```

```python
#Q6 convert a gray scale image into binary image without using built-in functions?
import cv2
import numpy as np

# Load the color image
image_path = r'C:\Users\2002a\Downloads\color.jpg'
color_image = cv2.imread(image_path)

# Check if image is loaded successfully
if color_image is None:
    print("Error: Unable to load image.")
else:
    # Convert the color image to grayscale
    gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    # Create a binary image with the same dimensions as the grayscale image
    binary_image = np.zeros_like(gray_image)

    # Define the threshold value
    threshold = 128

    # Iterate over each pixel and apply the threshold
    for i in range(gray_image.shape[0]):
        for j in range(gray_image.shape[1]):
            if gray_image[i, j] > threshold:
                binary_image[i, j] = 255
            else:
                binary_image[i, j] = 0

    # Resize images for display
    display_width = 400
    display_height = 300
    color_resized = cv2.resize(color_image, (display_width, display_height))
    gray_resized = cv2.resize(gray_image, (display_width, display_height))
    binary_resized = cv2.resize(binary_image, (display_width, display_height))

    # Create windows to display images
    cv2.namedWindow('Color Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Grayscale Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Binary Image', cv2.WINDOW_NORMAL)

    # Resize windows to the specified size (400x300)
    cv2.resizeWindow('Color Image', display_width, display_height)
    cv2.resizeWindow('Grayscale Image', display_width, display_height)
    cv2.resizeWindow('Binary Image', display_width, display_height)

    # Display the images
    cv2.imshow('Color Image', color_resized)
    cv2.imshow('Grayscale Image', gray_resized)
    cv2.imshow('Binary Image', binary_resized)



    # Wait for any key to be pressed to close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Q8 Manipulate colour spaces such as RGB, HSV ,LAB
import cv2
import numpy as np

# Load the color image
image_path = r'C:\Users\2002a\Downloads\color.jpg'
color_image = cv2.imread(image_path)

# Check if image is loaded successfully
if color_image is None:
    print("Error: Unable to load image.")
else:
    # Resize image for display
    display_width = 400
    display_height = 300
    resized_image = cv2.resize(color_image, (display_width, display_height))

    # Manipulate the RGB image by adding a constant value to each channel
    rgb_shift = np.array([20, 30, 10])
    rgb_result_image = np.clip(color_image.astype(int) + rgb_shift, 0, 255).astype(np.uint8)

    # Convert the color image to HSV
    hsv_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2HSV)

    # Manipulate the Hue channel by adding a constant value
    hue_shift = 20
    hsv_image[:, :, 0] = (hsv_image[:, :, 0] + hue_shift) % 180

    # Convert the color image to LAB
    lab_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2LAB)

    # Manipulate the a* channel by subtracting a constant value
    a_shift = 20
    lab_image[:, :, 1] = np.clip(lab_image[:, :, 1] - a_shift, 0, 255)

    # Convert back to BGR for display
    hsv_result_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
    lab_result_image = cv2.cvtColor(lab_image, cv2.COLOR_LAB2BGR)

    # Create windows to display images
    cv2.namedWindow('Original Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('RGB Manipulated Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('HSV Manipulated Image', cv2.WINDOW_NORMAL)
    cv2.namedWindow('LAB Manipulated Image', cv2.WINDOW_NORMAL)

    # Resize windows for better display
    cv2.resizeWindow('Original Image', display_width, display_height)
    cv2.resizeWindow('RGB Manipulated Image', display_width, display_height)
    cv2.resizeWindow('HSV Manipulated Image', display_width, display_height)
    cv2.resizeWindow('LAB Manipulated Image', display_width, display_height)

    # Display the images
    cv2.imshow('Original Image', resized_image)
    cv2.imshow('RGB Manipulated Image', rgb_result_image)
    cv2.imshow('HSV Manipulated Image', hsv_result_image)
    cv2.imshow('LAB Manipulated Image', lab_result_image)

    # Wait for any key to be pressed to close the windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Q9 Split and merge color channels in open cv
import cv2
import numpy as np

# Load the image
image_path = r'C:\Users\2002a\Downloads\kk.png'
image = cv2.imread(image_path)

# Resize the image to 400x300
image = cv2.resize(image, (400, 300))

# Split the image into its BGR channels
blue, green, red = cv2.split(image)

# Create binary masks for each color channel
threshold = 100  # Adjust this threshold to control what is considered as each color
red_mask = (red > threshold).astype(np.uint8) * 255
green_mask = (green > threshold).astype(np.uint8) * 255
blue_mask = (blue > threshold).astype(np.uint8) * 255

# Display only the areas corresponding to each color
red_area_only = cv2.bitwise_and(image, image, mask=red_mask)
green_area_only = cv2.bitwise_and(image, image, mask=green_mask)
blue_area_only = cv2.bitwise_and(image, image, mask=blue_mask)

# Show red areas in red color, green areas in green color, and blue areas in blue color
cv2.imshow('orignl image', image)
cv2.imshow('Red Areas', red_area_only)
cv2.imshow('Green Areas', green_area_only)
cv2.imshow('Blue Areas', blue_area_only)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
